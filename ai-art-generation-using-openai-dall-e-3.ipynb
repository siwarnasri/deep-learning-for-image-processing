# 1. Introduction

OpenAI, a pioneer in artificial intelligence research, is dedicated to ensuring that artificial general intelligence is beneficial to all of humanity. They launched a project named DALL-E in January 2021, which is a 12-billion parameter version of GPT-3. This AI model was trained to generate images from text descriptions, using a dataset of text-image pairs.

In 2022, OpenAI introduced DALL-E 2, an advanced AI system that can create realistic images and artwork from natural language descriptions. DALL-E 2 can generate original, lifelike images and art from a text prompt, combining various concepts, attributes, and styles.

Now, OpenAI has announced DALL-E 3, a significant advancement over its predecessors. DALL-E 3 is capable of understanding more nuance and detail, allowing for the translation of ideas into exceptionally accurate images1. It represents a leap forward in the ability to generate images that exactly adhere to the text provided1. Even with the same prompt, DALL-E 3 delivers significant improvements over DALL-E 2

We will first import all the necessary libraries required

```python
import os
import openai
import requests
from PIL import Image
from io import BytesIO

```

# Setting OpenAI API key

Next, we will define the key to be used to access the API
Go to Add-ons->Secrets and create key secret containing the actual key

```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("key")
print('complete!')
```

```python
from openai import OpenAI
client = OpenAI(api_key=secret_value_0)
```

# 2. Young Couple In Love

```python
response = client.images.generate(
  model="dall-e-3",
 prompt="impressionist painting of a young couple on a walk by a lake and in love",  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url

response = requests.get(image_url)
Image.open(BytesIO(response.content))
```
